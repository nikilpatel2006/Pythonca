```python
player1 = 'Mayank Yadav'
player2 = 'Mohsin Khan'
player3 = 'Ravi Bishnoi'

# Filter the data for the players
df_player1 = df[df['Player_Name'] == player1]
df_player2 = df[df['Player_Name'] == player2]
df_player3 = df[df['Player_Name'] == player3]
```

```python
# Story on the camparison on the performnce of Ravi Bishnoi, Mohsin Khan, & Mayank Yadav.
# Ravi Bishnoi is a better player , which have can also be seen in the ipl stats.
# Ravi had too much fluctuation in his ipl performance which make the team management confused
# wheather to take him in the game or not. But Mohsin had a direct downfall in his ipl career.
# And Mayank Yadav had never played the match, so there the no data to talk about him.
```

```python
# Plotting the performance of the players over the years

# Plot Runs Scored
plt.plot(df_player1['Year'], df_player1['Runs_Scored'], marker='o', label=f' {player1}')
plt.plot(df_player2['Year'], df_player2['Runs_Scored'], marker='o', label=f' {player2}')
plt.plot(df_player3['Year'], df_player3['Runs_Scored'], marker='o', label=f' {player3}')
plt.title('Comparison of IPL Performance Over the Years')
plt.xlabel('Year')
plt.ylabel('Runs Scored')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Balls bowled
plt.plot(df_player1['Year'], df_player1['Balls_Bowled'], marker='o', label=f' {player1}')
plt.plot(df_player2['Year'], df_player2['Balls_Bowled'], marker='x', label=f' {player2}')
plt.plot(df_player3['Year'], df_player3['Balls_Bowled'], marker='d', label=f' {player3}')
plt.title('Comparison of IPL Performance Over the Years')
plt.xlabel('Year')
plt.ylabel('Balls Bowled')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Catches Taken
plt.plot(df_player1['Year'], df_player1['Catches_Taken'], marker='d', linestyle='--', label=f' {player1}')
plt.plot(df_player2['Year'], df_player2['Catches_Taken'], marker='x', linestyle='--', label=f' {player2}')
plt.plot(df_player3['Year'], df_player3['Catches_Taken'], marker='o', linestyle='--', label=f' {player3}')
plt.title('Comparison of IPL Performance Over the Years')
plt.xlabel('Year')
plt.ylabel('Catches Taken')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Wickets Taken
plt.plot(df_player1['Year'], df_player1['Wickets_Taken'], marker='d', linestyle='--', label=f' {player1}')
plt.plot(df_player2['Year'], df_player2['Wickets_Taken'], marker='s', linestyle='-.', label=f' {player2}')
plt.plot(df_player3['Year'], df_player3['Wickets_Taken'], marker='s', linestyle='-.', label=f' {player3}')
plt.title('Comparison of IPL Performance Over the Years')
plt.xlabel('Year')
plt.ylabel('Wickets Taken')
plt.legend()
plt.grid(True)
plt.show()
```

```python
yearly_total_data = df_ipl.groupby('Year').agg({
    'Runs_Scored': 'sum',
    'Catches_Taken': 'sum',
    'Wickets_Taken': 'sum'
}).reset_index()

# Display the yearly summary
print(yearly_total_data)
```

```python
# story on overall ipl data from 2008 to 2024
# 2023 have been a most interesting year as the total run scored & total wickets taken are the highest and also
# the catches taken are good. The most boring year is 2008 where the all catches, wickets, runs are too low.
# the data show that the performance in ipl have been going better year by year, except 2017,2021,2024 where the performance have little degradation.
```

```python
# Plot Total Catches Taken Runs Scored

plt.plot(yearly_total_data['Year'], yearly_total_data['Runs_Scored'], marker='o', linestyle='-.', label= 'Runs Scored', color = 'red')
plt.title('Comparison of IPL Performance Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Total Catches Taken
plt.plot(yearly_total_data['Year'], yearly_total_data['Catches_Taken'], marker='s', linestyle='--', label='Catches Taken', color = 'purple')
plt.title('Comparison of IPL Performance Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plot Total Wickets Taken
plt.plot(yearly_total_data['Year'], yearly_total_data['Wickets_Taken'], marker='d', linestyle=':', label='Wickets Taken', color = 'green')
plt.title('Comparison of IPL Performance Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#All the wicket taken, run scored, and catches taken in one songle graph.

plt.plot(yearly_total_data['Year'], yearly_total_data['Wickets_Taken'], marker='d', linestyle=':', label='Wickets Taken', color = 'black')
plt.plot(yearly_total_data['Year'], yearly_total_data['Catches_Taken'], marker='s', linestyle='--', label='Catches Taken', color = 'yellow')
plt.plot(yearly_total_data['Year'], yearly_total_data['Runs_Scored'], marker='o', linestyle='-.', label= 'Runs Scored', color = 'red')
plt.title('Comparison of IPL Performance Over the Years')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.show()
```
